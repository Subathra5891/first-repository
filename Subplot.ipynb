## subplot
subplot(total number of rows,total number of columns,plot number(index))

eg: plt.subplot(1,2,1) will create a figure with one row and two column with first chart( axes 1)

plt.subplot(1,2,2) will create a chart in the axes 2

```python
#example for subplot ( one row and 3 column)
plt.subplot(1,3,1)
plt.title("axes 1") # title for first axes
plt.subplot(1,3,2)
plt.title("axes 2") # title for second axes
plt.subplot(1,3,3)
plt.title("axes 3") # title for 3rd axes
plt.suptitle("figure title") # title for the figure
plt.show()
```

```python
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.subplot(2,2,1)
sns.kdeplot(data=movie,x='CriticRatings',y='AudienceRatings', shade=True)
plt.subplot(2,2,2)
sns.kdeplot(data=movie,x='CriticRatings',y='AudienceRatings', shade=True)
plt.subplot(2,2,3)
sns.kdeplot(data=movie,x='CriticRatings',y='AudienceRatings', shade=True)
plt.subplot(2,2,4)
sns.boxplot(data=movie,x='Genre',y='AudienceRatings')
```

## it can also be created by

```python
sns.set_style('darkgrid')
fig,axes=plt.subplots(2,2,figsize=(15,15)) 
# plt.subplot always returns a tuple of figure and axes. so fig,axes are used to assign

k1=sns.boxplot(data=movie,x='Genre',y='AudienceRatings', ax=axes[0,0])

k2=sns.violinplot(data=movie,x='Genre',y='AudienceRatings',ax=axes[0,1])

k3=sns.stripplot(data=movie,x='CriticRatings',y='AudienceRatings',jitter=True,edgecolor='gray',size=8,palette='winter',
                 orient='v',ax=axes[1,0])

axes[1,1].hist(movie.CriticRatings,bins=100)
plt.show()
```
